```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
from bs4 import BeautifulSoup as bs
import pymysql
import time

# 1. 셀레니움으로 크롤링
options = Options()
options.binary_location = ".\\chrome-win64\\chrome.exe"
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920,1080")

service = Service("chromedriver.exe")
driver = webdriver.Chrome(service=service, options=options)
driver.get("https://www.starbucks.co.kr/index.do")

hover_element = driver.find_element(By.CSS_SELECTOR, "#gnb > div > nav > div > ul > li.gnb_nav03 > h2 > a")
menu_option = driver.find_element(By.CSS_SELECTOR, "#gnb > div > nav > div > ul > li.gnb_nav03 > div > div > div > ul:nth-child(1) > li.gnb_sub_ttl > a")
actions = ActionChains(driver)
actions.move_to_element(hover_element).perform()
time.sleep(1)
menu_option.click()
time.sleep(3)

# 지역검색 클릭
driver.find_element(By.CSS_SELECTOR, "#container > div > form > fieldset > div > section > article.find_store_cont > article > header.loca_search > h3 > a").click()
time.sleep(1)
driver.find_element(By.CSS_SELECTOR, "#container > div > form > fieldset > div > section > article.find_store_cont > article > article:nth-child(4) > div.loca_step1 > div.loca_step1_cont > ul > li:nth-child(6) > a").click()
time.sleep(1)
driver.find_element(By.CSS_SELECTOR, "#mCSB_2_container > ul > li:nth-child(1) > a").click()
time.sleep(3)

html = driver.page_source
driver.quit()

# 2. BeautifulSoup으로 데이터 파싱
soup = bs(html, 'html.parser')
contents = soup.select("#mCSB_3_container > ul > li")

star_list = []
for item in contents:
    try:
        store_name = item['data-name']
        lat = item['data-lat']
        lng = item['data-long']
        address = item.select_one('p.result_details').text.strip()
        store_type = item.select_one('i').get('class')[0]

        star_list.append((store_name, address, lat, lng, store_type))
    except Exception as e:
        print(f"Error parsing item: {e}")

print(f"✅ 총 {len(star_list)}개 매장 크롤링 완료!")

# 3. MySQL: 데이터베이스 생성
try:
    conn = pymysql.connect(
        # host="210.119.14.56",
        user="root",
        password="1234",
        port=3306,
        charset="utf8mb4"
    )
    cursor = conn.cursor()
    cursor.execute("CREATE DATABASE IF NOT EXISTS JS6881 CHARACTER SET utf8mb4 COLLATE utf8mb4_general_ci;")
    print("✅ 데이터베이스 'JS6881' 생성 완료!")
    cursor.close()
    conn.close()

except Exception as e:
    print("❌ 데이터베이스 생성 중 오류 발생:", e)

# 4. MySQL: 데이터 저장
try:
    conn = pymysql.connect(
        host="210.119.14.56",
        user="root",
        password="1234",
        database="JS6881",  # ✅ 생성한 DB에 연결
        port=3306,
        charset="utf8mb4"
    )
    print("✅ JS6881 DB에 연결 성공!")
    cursor = conn.cursor()

    # 테이블 생성
    cursor.execute("DROP TABLE IF EXISTS KJS")
    cursor.execute("""
        CREATE TABLE KJS (
            store_name VARCHAR(100),
            address VARCHAR(300),
            latitude VARCHAR(50),
            longitude VARCHAR(50),
            DT_y_n VARCHAR(50)
        )
    """)
    print("✅ 테이블 KJS 생성 완료!")

    # 데이터 삽입
    if not star_list:
        print("❌ star_list가 비어 있습니다.")
    else:
        insert_sql = "INSERT INTO KJS (store_name, address, latitude, longitude, DT_y_n) VALUES (%s, %s, %s, %s, %s)"
        cursor.executemany(insert_sql, star_list)
        conn.commit()
        print("✅ MySQL DB 저장 완료!")

    cursor.close()
    conn.close()

except Exception as e:
    print("❌ MySQL DB 처리 중 오류 발생:", e)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
from bs4 import BeautifulSoup as bs
import oracledb
import time

# 1. 셀레니움으로 크롤링
options = Options()
options.binary_location = ".\\chrome-win64\\chrome.exe"
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920,1080")

service = Service("chromedriver.exe")
driver = webdriver.Chrome(service=service, options=options)
driver.get("https://www.starbucks.co.kr/index.do")

hover_element = driver.find_element(By.CSS_SELECTOR, "#gnb > div > nav > div > ul > li.gnb_nav03 > h2 > a")
menu_option = driver.find_element(By.CSS_SELECTOR, "#gnb > div > nav > div > ul > li.gnb_nav03 > div > div > div > ul:nth-child(1) > li.gnb_sub_ttl > a")
actions = ActionChains(driver)
actions.move_to_element(hover_element).perform()
time.sleep(1)
menu_option.click()
time.sleep(3)

# 지역검색 클릭
driver.find_element(By.CSS_SELECTOR, "#container > div > form > fieldset > div > section > article.find_store_cont > article > header.loca_search > h3 > a").click()
time.sleep(1)
driver.find_element(By.CSS_SELECTOR, "#container > div > form > fieldset > div > section > article.find_store_cont > article > article:nth-child(4) > div.loca_step1 > div.loca_step1_cont > ul > li:nth-child(6) > a").click()
time.sleep(1)
driver.find_element(By.CSS_SELECTOR, "#mCSB_2_container > ul > li:nth-child(1) > a").click()
time.sleep(3)

html = driver.page_source
driver.quit()

# 2. BeautifulSoup으로 데이터 파싱
soup = bs(html, 'html.parser')
contents = soup.select("#mCSB_3_container > ul > li")

star_list = []
for item in contents:
    try:
        store_name = item['data-name']
        lat = item['data-lat']
        lng = item['data-long']
        address = item.select_one('p.result_details').text.strip()
        store_type = item.select_one('i').get('class')[0]

        star_list.append((store_name, address, lat, lng, store_type))
    except Exception as e:
        print(f"Error parsing item: {e}")

print(f"✅ 총 {len(star_list)}개 매장 크롤링 완료!")

# 3. Oracle DB에 저장
try:
    conn = oracledb.connect(
        user="PKNU",
        password="1234",
        host="localhost",
        port=1521,
        service_name="XEPDB1"
    )
    print("✅ Oracle 연결 성공!")

    cursor = conn.cursor()

    # 기존 테이블 삭제 (있다면)
    cursor.execute("""
    BEGIN
        EXECUTE IMMEDIATE 'DROP TABLE starbucks_busan';
    EXCEPTION
        WHEN OTHERS THEN NULL;
    END;
    """)

    # 테이블 생성
    cursor.execute("""
    CREATE TABLE starbucks_busan (
        store_name VARCHAR2(100),
        address VARCHAR2(300),
        latitude VARCHAR2(50),
        longitude VARCHAR2(50),
        store_type VARCHAR2(50)
    )
    """)

    # 데이터 삽입
    if not star_list:
        print("❌ star_list가 비어 있습니다.")
    else:
        cursor.executemany(
            "INSERT INTO starbucks_busan (store_name, address, latitude, longitude, store_type) VALUES (:1, :2, :3, :4, :5)",
            star_list
        )
        conn.commit()
        print("✅ DB 저장 완료!")

    cursor.close()
    conn.close()

except Exception as e:
    print("❌ DB 처리 중 오류 발생:", e)
```
